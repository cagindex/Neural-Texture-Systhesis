```python
from pathlib import Path
from PIL import Image
import os
import numpy as np
```

```python
name = 'strawberry.jpg'
PATH  = (Path(os.path.abspath('')) / 'Images' / name).resolve()
SPATH = (Path(os.path.abspath('')) / 'Saves' / name).resolve()
```

```python
image = Image.open(PATH)
image
```

```python
def cut(data : np.ndarray, x : int, y : int, width : int, height : int) :
    return data[x : x + width, y : y + height, :]
```

```python
def show(data : np.ndarray):
    return Image.fromarray(data)
```

```python
data = np.array(image)
data.shape
```

```python
show(cut(data, 0, 0, 394, 394)).save(SPATH)
```
